## Simple dataset 
* keep only features related to salary
* wage per hour, weeks worked in year, capital gains, capital losses, dividends from stocks

```python
keep_columns = ["wage per hour", "weeks worked in year", "capital gains", "capital losses", "dividends from stocks"]
X_simple_df = df[keep_columns]
Y_simple_df = pd.DataFrame(df["income class"] == ' 50000+.').astype('int')
```

```python
from sklearn import preprocessing

X_simple = X_simple_df.values
min_max_scaler = preprocessing.MinMaxScaler()
X = min_max_scaler.fit_transform(X_simple)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y_simple_df, test_size=0.3, random_state=42)
```

```python
lr_predictor = LogisticRegression(random_state=0).fit(X_train, Y_train)
```

### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier

rf_predictor = RandomForestClassifier(random_state=0).fit(X_train, Y_train)
```

### Validation

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(Y_test, lr_predictor.predict(X_test))
```

```python
roc_auc_score(Y_test, rf_predictor.predict(X_test))
```
